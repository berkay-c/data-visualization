```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('./homework_data.csv')
```

```python
df
```

```python
import matplotlib.dates as mdates
myFmt = mdates.DateFormatter('%d')

```

```python
colors = ["#215968", "#31859C", "#FFC000"]
labels = ["International", "Services", "Domestic"]
sns.set_style("white")
plt.figure(figsize = (12, 6))
plt.stackplot(df["Date"],
              df["International"],
              df["Services"],
              df["Domestic"],
              colors = colors,
              edgecolor = 'none')
plt.xticks(df["Date"], rotation = 45)
plt.legend(labels = labels, loc = "upper left")
plt.ylabel("Revenue in $", fontsize = 13)
plt.title("Revenue Report 2020", fontsize = 14, weight = "bold")
sns.despine()
plt.show()
```
