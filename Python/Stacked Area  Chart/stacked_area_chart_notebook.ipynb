```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_fuel_engine_types = pd.read_csv('./stacked_area_chart_data.csv')
```

```python
df_fuel_engine_types.head(6)
```

```python
color= ["#011638","#7e2987","ef2026"]
sns.set_style('white')
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_types['Year'],
             df_fuel_engine_types['Diesel'],
             df_fuel_engine_types['Petrol'],
             df_fuel_engine_types['Gas'],
             colors=color,
             edgecolor='none')
plt.xticks(df_fuel_engine_types['Year'],rotation=45,fontsize=11,fontweight='bold')
plt.yticks(fontsize=14,fontweight='bold')
plt.legend(["Diesel","Petrol","Gas"],loc="upper left")
plt.ylabel('Number of Cars',fontsize=14,fontweight='bold')
plt.title('Popularity of Engine Fuel Types (1982-2016)',fontsize=19,fontweight='bold')
sns.despine()
plt.savefig('stacked_area_chart_image.png',bbox_inches='tight')
plt.show()
```
