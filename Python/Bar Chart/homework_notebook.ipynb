```python
import pandas  as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('./bar_chart_homework_data.csv')
data
```

```python
plt.figure(figsize=(9,6))
plt.barh(y = data['Cities'],
        width= data['Frequency'],
       color = "midnightblue")
plt.xticks(fontsize=13)
plt.yticks(rotation=45,fontsize=13)
plt.title('Frequency of Cities',fontsize=16,fontweight='bold')
plt.xlabel('Number of Frequency', fontsize =13)
plt.savefig("bar_chart_homework_image.png",bbox_inches='tight')
plt.show()
```
