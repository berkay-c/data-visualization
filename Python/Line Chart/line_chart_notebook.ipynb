```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('./line_chart_data.csv')
```

```python
df['New Date'] = pd.to_datetime(df['Date'])
```

```python
df
```

```python
last_quarter=df[(df['New Date'] >= '2008-07-01') &
                (df['New Date'] <= '2008-12-31')]
```

```python
last_quarter
```

```python
label=['S&P 500','FTSE 100']
plt.figure(figsize=(20,6))
plt.plot(last_quarter['New Date'],last_quarter['GSPC500'],color='midnightblue')
plt.plot(last_quarter['New Date'],last_quarter['FTSE100'],color='crimson')
plt.title('S&P vs FTSE Returns (2000-2010)',fontsize=18,fontweight='bold')
plt.ylabel('Returns',fontsize=14,fontweight='bold')
plt.xlabel('Date',fontsize=14,fontweight='bold')
plt.legend(labels=label,fontsize=12)
plt.show()
```
